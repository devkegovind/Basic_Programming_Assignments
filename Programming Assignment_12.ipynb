**1. Write a Python program to Extract Unique values dictionary values?**


```python
# Python program to extract Unique values from dictionary values

# initialize the dictionary

dict_1 = {'gfg': [5, 6, 7, 8],
             'is': [10, 11, 7, 5],
             'best': [6, 12, 10, 8],
             'for': [1, 2, 5]}

print(f"Original dictionary is {dict_1}.")

list_1 = []
for value in dict_1.values():
  for ele in value:
    list_1.append(ele)
print(sorted(list(set(list_1))))
```

**2. Write a Python program to find the sum of all items in a dictionary?**


```python
# Python program to find the sum of all items in a dictionary

def sum_of_items(dict):
  list_2 = []

  for i in dict:
    list_2.append(dict[i])
  return sum(list_2)

dict_1 = {'gvd': 500, 'dgv': 250, 'ngd': 750}
sum_of_items(dict_1)
print(f"The sum of all items in a dictionary is {sum_of_items(dict_1)}.")
```

**3. Write a Python program to Merging two Dictionaries?**


```python
# Python program to merging two dictionaries

def merge(dict1, dict2):
  for i in dict2.keys():
    dict1[i] = dict2[i]
  return dict1


dict_1 = {1: 'a', 2: 'b'}
dict_2 = {2: 'c', 4: 'd'}

print(f"The merging two dictionaries is {merge(dict_1, dict_2)}.")
```

**4. Write a Python program to convert key-values list to flat dictionary?**


```python
# Python program to convert key-values list to flat dictionary

def flat_dict(dict):
  return list(dict.keys())

students = {
  'Theodore': 19,
  'Roxanne': 20,
  'Mathew': 21,
  'Betty': 20
}
print(f"convert key-values list to flat dictionary is {flat_dict(students)}.")
```

**5. Write a Python program to insertion at the beginning in OrderedDict?**

```python
# Python program to insertion at the begining in OrderedDict

"""
Input: 
original_dict = {'a':1, 'b':2}
item to be inserted ('c', 3)

Output:  {'c':3, 'a':1, 'b':2}

Input: 
original_dict = {'akshat':1, 'manjeet':2}
item to be inserted ('nikhil', 3)

Output:  {'nikhil':3, 'akshat':1, 'manjeet':2}
"""
from collections import OrderedDict
 # initialize ordered_dict
inordered_dict = OrderedDict([('akshay', '1'), ('nikhil', '2')])

 # inserting items in starting of dict
inordered_dict.update({'manoj': '3'})
inordered_dict.move_to_end('manoj', last = False)

print(f"The result is {inordered_dict}")
```

**6. Write a Python program to check order of character in string using OrderedDict()?**


```python
# Python program to check order of character in string using OrderedDict()
import collections

ordered_dict = collections.OrderedDict()

ordered_dict['1'] = "one"
ordered_dict['2'] = "two"
ordered_dict['3'] = 'three'
ordered_dict['4'] = 'four'

print("Printing Ordered Dictionary: ")

for key, value in ordered_dict.items():
  print(f"key : {key}, value : {value}.")
```
