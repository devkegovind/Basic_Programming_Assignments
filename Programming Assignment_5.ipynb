**1. Write a Python Program to Find LCM?**
- The least common multiple(L.C.M.) of two numbers is the smallest positive integer that is perfectly divisible by the two given numbers.
- For example, the LCM of 12 and 14 is 84.




```python
# Python Program to find LCM
def lcm(x, y):
  # choose the greater number
  if x > y:
    greater = x
  else:
    greater = y

  while(True):
    if ((greater % x == 0) and (greater % y == 0)):
      lcm = greater
      break
    greater += 1

  return lcm

num1 = 54
num2 = 24

print(f"The LCM is: {lcm(num1, num2)}")
```

**2. Write a Python Program to Find HCF?**
- The highest common factor(H.C.F.) or greatest common divisior(G.C.D.) of two numbers is the largest positive integer that perfectly divides the two given numbers. 
- For example, the HCF of 12 and 14 is 2.

```python
# Python program to Find HCF

def hcf(x, y):

  # choose the smaller number
  if x > y:
    smaller = y
  else:
    smaller = x

  for i in range(1, smaller + 1):
    if ((x % i == 0) and (y % i == 0)):
      hcf = i
  return hcf

num1 = 54
num2 = 24

print(f"The HCF is:{hcf(num1, num2)}")
```

**5. Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?**

```python
# Python Program to Make a Simple Calculator

x = int(input("Enter the value1: "))
y = int(input("Enter the value2: "))

print(f"The addition of two numbers is:{x + y}")
print(f"The substraction of two numbers is:{x - y}")
print(f"The multiplication of two numbers is:{x*y}")
print(f"The division of two numbers is:{x /y}")
```

```python
# Program make a simple calculator

# This function adds two numbers
def add(x, y):
  return x + y


# This function subtracts two numbers 
def subtract(x, y):
  return x - y


# This function multiples two numbers
def multiple(x, y):
  return x * y

# This function divides two numbers
def divide(x, y):
  return x / y

print("Select Opertaion")
print("1.Add")
print("2.Subtract")
print("3.Multiply")
print("4.Divide")

while True:
  # take input from the user 
  choice = input("Enter choice(1/2/3/4): ")

  # check if choice is one of the four options
  if choice in ('1', '2', '3', '4'):
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    if choice == '1':
      print(f"Addition of two numbers is:{add(num1, num2)}")

    elif choice == '2':
      print(f"Substraction of two numbers is: {subtract(num1, num2)}")

    elif choice == '3':
      print(f"Multiplication of two numbers is:{multiple(num1, num2)}")

    elif choice == 4:
      print(f"Division of two numbers is:{divide(num1, num2)}")

    # check if user wants another calculation
    # check the while loop if answer is no
    next_calculation = input("Let's do next calculation? (yes/no): ")
    if next_calculation == 'no':
      break
  else:
      print("invalid")
```
