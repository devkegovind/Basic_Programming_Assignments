**1. Write a Python Program to Add Two Matrices?**


```python
# Python Program to Add Two Matrices Using Nested Loop

X = [[1, 2, 3],
     [4, 5, 6],
     [7, 8, 9]]

Y = [[10, 11, 12],
     [13, 14,15],
     [16, 17, 18]]

XY = [[0, 0, 0],
      [0, 0, 0],
      [0, 0, 0]]

# iterate rows first from X matrix
for i in range(len(X)):
  # iterate columns from X matrix
  for j in range(len(X[0])):
    XY[i][j] = X[i][j] + Y[i][j]

for i in XY:
  print (i)
```

**3. Write a Python Program to Transpose a Matrix?**


```python
# Python Program to Transpose a Matrix

a = [[10, 20],
     [30, 40],
     [50, 60]]

result = [[0, 0, 0],
          [0, 0, 0]]

# iterate rows from a matrix
for i in range(len(a)):

  # iterate columns from a matrix
  for j in range(len(a[0])):

    result[j][i] = a[i][j]

for i in result:
  print(i)
```

**4. Write a Python Program to Sort Words in Alphabetic Order?**


```python
# Python Program to sort words in Alphabetic Order

my_words = "Hello im Devke Govind from Pune."

# Breakdown the string into list of words
words = [ word.lower() for word in my_words.split()]

# sort the words
words.sort()

# display the word
for word in words:
  print(word)
```

**5. Write a Python Program to Remove Punctuation From a String?**

```python
# define punctuation

punctuation = """!@#$%^&*()_-+={}[];:,./?><`~"""

my_string = "Hello!!!, he said --- and went."

# remove punctuation from string

# empty string
no_punct = ""

for char in my_string:
  if char not in punctuation:
    no_punct = no_punct + char
print(no_punct)
```
