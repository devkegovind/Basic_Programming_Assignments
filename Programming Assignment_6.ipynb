**1. Write a Python Program to Display Fibonacci Sequence Using Recursion?**
- **Fibonacci sequence:**
A fibonacci sequence is a sequence of intergers which first two terms are 0 and 1 and all other terms of the sequence are obtained by adding their preceding two numbers.
- For example: 0,1,1,2,3,5,8,13 and so on.


```python
def recur_fibo(n):
  if n <= 1:
    return n
  else:
    return(recur_fibo(n-1) + recur_fibo(n-2))

# take input from the user
nterms = int(input("Enter number of terms: "))

# check if the number of terms is valid
if nterms <= 0:
  print("Please enter a positive integer.")
else:
  print("Fibonacci sequence")

  for i in range(nterms):
    print(recur_fibo(i))
```

**2. Write a Python Program to Find Factorial of Number Using Recursion?**
- The factorial of a number is the product of all the integers from 1 to that number.
- for example, the factorial of 6 is `1*2*3*4*5*6 = 720`.
- Factorial is not defined for negative numbers and the factorial of zero is one, 0! = 1.

```python
def recur_factorial(n):
  if n == 1:
    return n
  else:
    return n * recur_factorial(n-1)

num = 7
 # check if the number is negative
if num < 0:
   print(f"The give number{num} factorial does not exists.")
elif num == 0:
  print(f"The given number{num} factorial is 1.")
else:
  print(f"The given number{num} factorial is {recur_factorial(num)}")
```

**3. Write a Python Program to calculate your Body Mass Index?** 
- **BMI** short for **Body Mass Index**, is a measure of relative weight based on the mass and height of an individual. We generally use the Body Mass Index in order to categorize people on the basis of their **height** and **weight**. These categories are **underweight, healthy, overweight**, and even **obesity**. 


```python
# Python prgram to calculate your Body Mass Index
height = float(input("Enter the height in cm: "))
weight = float(input("Enter the weight in Kg: "))
BMI = weight/(height/100)**2
print(f"Your Body Mass Index is {BMI}.")

# Using the if-elif-else conditions
if BMI <= 18.5:
  print("Oops! You are underweight.")
elif BMI <= 24.9:
  print("Awesome! You are healthy.")
else:
  print("Seesh! You are obese.")
```

**5. Write a Python Program for cube sum of first n natural numbers?**
- Natural numbers are a **part of the number system which includes all the positive integers from 1 till infinity and are also used for counting purpose.** It does not include zero(0).
- It is also called counting numbers.

Input: n = 5

Output: 225

$1^3 + 2^3 + 3^3 + 4^3 + 5^3 = 225$

```python
def sumofSeries(n):
  sum = 0
  for i in range(1, n+1):
    sum = sum + i**3

  return sum

n = 5
print(f"The cube sum of first {n} natural numbers is {sumofSeries(n)}.")
```
