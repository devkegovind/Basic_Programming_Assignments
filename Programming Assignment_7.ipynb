**1. Write a Python Program to find sum of array?**
- Given an array of integers, find the sum of its elements. 

Input : arr[] = {1, 2, 3}

Output: 6

$1 + 2 + 3 = 6$

```python
# Iterating through the array and adding each element to the sum variable and finally displaying the sum.

def _sum(arr):

  # initialize a variable
  # to store the sum
  # while iterating through
  # the array later
  sum = 0

  # iterate through the array
  # and add each element to the sum variable
  # one at a time
  for i in arr:
    sum = sum + i

  return (sum)


arr = []
# input values to list
arr = [12, 3, 4, 15]

# calculating length of array
n = len(arr)

ans = _sum(arr)

print(f"Sum of array is {ans}.")

```

Using the `reduce` method. Array.reduce() method is used to iterate over the array and get the summarized result from all elements of array.
- `reduce(function, array)`

```python
from functools import reduce


def _sum(arr):


  sum = reduce(lambda a, b: a+b, arr)

  return (sum)

arr = [12, 3, 4, 15]

n = len(arr)

ans = _sum(arr)

print(f"The sum of array is {ans}.")
```

**2. Write a Python Program to find largest element in an array?**

```python
def lar_ele(arr, n):
  max = arr[0]

  for i in range(1, n):
    if arr[i] > max:
      max = arr[i]
  return max

arr = [12, 3, 4, 15]
n = len(arr)

print(f"The largest element in an array is {lar_ele(arr, n)}.")
```

**Using sort() function:**

```python
def largest(arr, n):

  # sort the array

  arr.sort()

  # The last element of the array is the largest element

  return arr[-1]

arr = [10,324, 45, 90, 9808]

n = len(arr)

print(f"The largest element in an array is {largest(arr, n)}.")
```

**3. Write a Python Program for array rotation?**

**Time Complexicity:O(n)**

**Auxiliary Space:O(1)**

**Approach:** Using List slicing

```python
# Python Program for array rotation
def rotateList(arr, d, n):
  arr[:] = arr[d:n] + arr[0:d]

  return arr

arr = [1, 2, 3, 4, 5, 6]

print(f"The given arr is {arr}.")

print(f"The array rotation is {rotateList(arr, 2, len(arr))}")
```

**4. Write a Python Program to Split the array and add the first part to the end?**

```python
# Input : arr[] = {12, 10, 5, 6, 52, 36}
          # k = 2
# output : arr[] = {5, 6, 52, 36, 1, 2}
# split from index 2 and first part {12, 10} add to the end.

def splitArr(arr, n, k):
  for i in range(0, k):
    x = arr[0]
    for j in range(0, n-1):
      arr[j] = arr[j + 1]

    arr[n-1] = x

arr = [12, 10, 5, 6, 52, 36]

n = len(arr)

position = 2

splitArr(arr, n, position)

for i in range(0, n):
  print(arr[i], end = " ")
```


**5. Write a Python Program to check if given array is Monotonic?**
- To check whether an array A is monotone increasing, we'll check **A[i] <= A[i+1] for all i indexing from 0 to len(A)-2**. Similarly we can check for monotone decreasing where A[i]>= A[i+1] for all i indexing from 0 to len(A)-2.

```python
# Python Program to check if given array is Monotonic

# check if given array is Monotonic

def isMonotonic(A):
  return (all(A[i] <= A[i + 1] for i in range(len(A)- 1)) or
          all(A[i] >= A[i + 1] for i in range(len(A) -1)))
  
A = [6, 5, 4, 4]

print(f"The array is Monotonic {isMonotonic(A)}.")
```
