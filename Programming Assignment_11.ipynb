**1. Write a Python program to find words which are greater than given length k?**


```python
# Python program to find words which are greater than given length k

def findwords(str, k):
  
  # create empty string
  str1 = []

  # split the given string

  text = str.split(" ")

  # iterate the string
  for i in text:
    if len(i) > k:
      str1.append(i)
  return str1

str2 = "hello geeks for geeks is computer science portal" 
k = 4
print(f"The words which are greater than given length k is {findwords(str2, k)}.")
```

**3. Write a Python program to split and join a string?**


```python
# Python program to split and join a string

def split_string(string):

  # split the string based on space delimiter
  list_string = string.split(" ")

  return list_string

def join_string(list_string):

  # join the string based on "_" delimiter

  string = '-'.join(list_string)

  return string

string = "Devke Govind Vaijanath"
list_string = split_string(string)
print(f"The splited string is {list_string}.")

# join list of strings into one
new_string = join_string(list_string)
print(f"The joined new string is {new_string}.")
```

**4. Write a Python to check if a given string is binary string or not?**


```python
# Python program to check if a given string is binary string or not

def check(string):

  # set function convert string into set of characters

  p = set(string)

  # declare set of '0', '1'
  s = {'0', '1'}

  # check set p is same as set s
  # or set p contains only '0'
  # or set p contains only '1'
  # or not, if any one condition
  # is true then string is accepted
  # otherwise not.
  if s == p or p == {'0'} or p == {'1'}:
    print("Yes")
  else:
    print("No")

string = "101010000111"

check(string)
```

**5. Write a Python program to find uncommon words from two Strings?**


```python
"""
Input : A = "apple banana mango" 
        B = "banana fruits mango"
Output : ['apple', 'fruits']

"""
A = "apple banana mango"

B = "banana fruits mango"

c = []

for i in A.split():
  for j in B.split():
    if i not in B.split() and j not in A.split():
      c.append(i)
      c.append(j)
print(c)
    
```

**6. Write a Python to find all duplicate characters in string?**


```python
# Python to find all duplicate character in string
def find_dup_char(string):
  x = []

  for i in string:
    if i not in x and len(string) > 1:
      x.append(i)
  print(" ".join(x))

string = "Engineering"
find_dup_char(string)
```
