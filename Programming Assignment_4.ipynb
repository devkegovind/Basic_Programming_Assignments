**1. Write a Python Program to Find the Factorial of a Number?**

```python
# Python Program to Find the Factorial of a Number

num = int(input("Enter the number: "))

factorial = 1

if num < 0:
  print(f"The factorial of given number {num} does not exist!.")
elif num == 0:
  print(f"The factorial of given number {num} is 1.")
else:
  for i in range(1, num + 1):
    factorial = factorial * i
  print(f"The factorial of given number {num} is {factorial}.")
```

**3. Write a Python Program to Print the Fibonacci sequence?**
- Fibonacci Sequence specifies a series of numbers where the next number is found by adding up the two numbers just before it.


```python
# Python Program to print the Fibonacci sequence

num = int(input("Enter the number: "))

count = 0
n_1 = 0
n_2 = 1

if num <= 0:
  print(f"The given number{num} it is not a valid number for series.")
elif num == 1:  
  print(f"The fibonacci sequence of the given number{num} is {n_1}.")
else:
  print(f"The fibonacci sequence of the given number{num} is:")
  while count < num:
    print(n_1)
    nth = n_1 + n_2
    n_1 = n_2
    n_2 = nth
    count += 1
```

**4. Write a Python Program to Check Armstrong Number?**

- Armstrong Number is defined as a number that consist of three such digits sum of the third power is equal to the original number.

```python
# Python program to check armstrong number

# take input from the user
num = int(input("Enter a number: "))

# initialize sum
sum = 0

# find the sum of the cube of each digit
temp = num
while temp > 0:
  digit = temp % 10
  sum = sum + digit ** 3
  temp = temp // 10

# display the result
if num == sum:
  print(f"The given number{num} is an Armstrong number")
else:
  print(f"The given number{num} is not an Armstrong number.")
```

**5. Write a Python Program to Find Armstrong Number in an Interval?**
